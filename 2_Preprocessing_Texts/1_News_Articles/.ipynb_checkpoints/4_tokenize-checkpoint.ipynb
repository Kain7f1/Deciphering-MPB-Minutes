## 4-c. 토큰화 테스트

### 4-c-1. 테스트 : ekonlpy를 이용한 토큰화

```python
from ekonlpy import Mecab

# 1. 객체 선언
mecab = Mecab(use_original_tagger=True) # set use_original_tagger=True

# 2. mecab.pos() 로 text를 토큰화
tokens_temp = mecab.pos(text_temp)
type(tokens_temp)
```

```python
tokens_temp
```

### 4-c-2. 품사 제한
- 명사, 형용사, 부사, 동사 만 추출한다
- 'NNG', 'NNP', 'VA', 'VAX', 'MAG', 'MAJ', 'VV', 'VV+EF', 'VV+EC', 'VV+ETN', 'VV+ETM'

```python
# 함수 : 토큰 리스트를 받아, 품사를 거르는 함수
def filter_pos_tag(tokens):
    
    # 유지하고 싶은 품사 태그
    desired_tags = ['NNG', 'NNP', 'VA', 'VAX', 'MAG', 'MAJ', 'VV', 'VV+EF', 'VV+EC', 'VV+ETN', 'VV+ETM']
    
    # 필터링된 리스트 생성
    filtered_tokens = [item for item in tokens if item[1] in desired_tags]

    return filtered_tokens
```

```python
filter_pos_tag(tokens_temp)
```

### 4-c-3. 토큰화 전 과정 함수화

```python
from ekonlpy import Mecab


def tokenize_using_ekonlpy(text):
    
    # 1. 객체 선언
    mecab = Mecab(use_original_tagger=True)
    
    # 2. mecab.pos() 로 text를 토큰화
    tokens = mecab.pos(text)

    # 3. 유지하고 싶은 품사 태그
    desired_tags = ['NNG', 'NNP', 'VA', 'VAX', 'MAG', 'MAJ', 'VV', 'VV+EF', 'VV+EC', 'VV+ETN', 'VV+ETM']
    
    # 4. 필터링된 리스트 생성
    filtered_tokens = [item for item in tokens if item[1] in desired_tags]

    return filtered_tokens
```

```python
text_temp = df.iloc(0)[1]["text"]
text_temp
```

```python
tokenize_result = tokenize_using_ekonlpy(text_temp)
tokenize_result
```

## 4-d. 함수화
- 토크나이즈 전 과정을 함수화한다.

### 4-d-1. 함수 정의

```python
from ekonlpy import Mecab
import pandas as pd
import time

# 시간 측정 데코레이터
def time_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Function '{func.__name__}' executed in {end_time - start_time} seconds.")
        return result
    return wrapper

# 함수 : 토큰화 + 파일로 저장
@time_decorator
def tokenize_csv_file(fpath_to_read, text_column_name, fpath_to_save):
    
    # 1. 데이터 읽어오기
    print("[progress 1/4] 데이터 읽어오기", end="\n\n")
    df = pd.read_csv(fpath_to_read, encoding="utf-8")
    print(df.info())


    # 2. 함수 정의 : 토큰화
    print("[progress 2/4] 함수 정의 : 토큰화", end="\n\n")
    def tokenize_using_ekonlpy(text):
        mecab = Mecab(use_original_tagger=True)   # 1. 객체 선언
        tokens = mecab.pos(text)   # 2. mecab.pos() 로 text를 토큰화
        desired_tags = ['NNG', 'NNP', 'VA', 'VAX', 'MAG', 'MAJ', 'VV', 'VV+EF', 'VV+EC', 'VV+ETN', 'VV+ETM']   # 3. 유지하고 싶은 품사 태그
        filtered_tokens = [item for item in tokens if item[1] in desired_tags]   # 4. 필터링된 리스트 생성
    
        return filtered_tokens

    # 3. df에 토큰화 적용
    print("[progress 3/4] df에 토큰화 적용", end="\n\n")
    df["tokens"] = df["text"].apply(tokenize_using_ekonlpy)
    print(df.head(3))

    # 4. 파일로 저장 : 토큰화된 데이터
    print("[progress 4/4] 파일로 저장 : 토큰화 적용된 데이터", end="\n\n")
    df_new = df.drop([text_column_name], axis=1)  # 불필요한 칼럼 제거
    df_new.to_csv(fpath_to_save, encoding='utf-8', index=False)  # csv 파일로 저장 

    print("[done]")
```
